```python
import pandas as pd
```

```python
data_path = '../data/'
```

## Examples for creating samples for Lac1

```python
# For CSV data
df1 = pd.read_csv(data_path+'Lac1_total.csv')
print(df1)
```

```python
ind_dg = df1['vr'] <= -160
```

```python
dg = df1[ind_dg]
dg
```

```python
dg.to_csv(data_path+'Lac1_dg1.csv')
```

```python
sample1 = df1.sample(n=int(len(df1)*0.5))
sample1
```

```python
sample2 = df1.sample(n=int(len(df1)*0.1))
sample2
```

```python
sample1.to_csv(data_path+'Lac1_dg_s1.csv')
sample2.to_csv(data_path+'Lac1_dg_s2.csv')
```
