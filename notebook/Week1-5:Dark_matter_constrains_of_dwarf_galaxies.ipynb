```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
data_path = '../data/'
```

```python
df = pd.read_csv(data_path+'Cas3_dg.csv')

rh = np.loadtxt(data_path+'rh_chains/CasIII_rh_chain.txt')
print (len(rh))
chain = np.loadtxt(data_path+'chain_Cas3.txt')
print (chain.shape)
```

Distance modulus:

Cas III:   24.45+/-0.14

Lac I:     24.40+/-0.12

Per I:     24.49+/-0.18

dist_modulus = 5log_10(D(pc)) - 5

rh (arcmin) -> rh (pc)

M(Msun) = 580 x rh(pc) x sigma_{vr}^2

Cas III: Mv = -12.3+- 0.7

Lv/Lv,sun = 10**(-0.4)(Mv-4.83)



```python
dm_cas3 = np.random.normal(24.45, 0.14, 100000)

d_cas3 = 10**((dm_cas3+5)/5)

plt.figure()
plt.hist(d_cas3, bins=50, alpha=0.5, label='d_cas3');
plt.xlabel("D (pc)", fontsize=20)
```

```python
dvr_cas3 = np.random.choice(chain[:,1], size=100000)
rh_cas3 = np.random.choice(rh, size=100000)/60/180*np.pi

plt.figure()
plt.hist(rh_cas3, bins=50, alpha=0.5, label='rh_cas3');
plt.xlabel("rh (arcmin)", fontsize=20)


plt.figure()
plt.hist(d_cas3*np.tan(rh_cas3), bins=50, alpha=0.5, label='rh_cas3');

plt.xlabel("rh (pc)", fontsize=20)
```

```python
mtot_cas3 = 580*d_cas3*np.tan(rh_cas3)*dvr_cas3**2.

plt.figure()
plt.hist(mtot_cas3, bins=50, alpha=0.5, label='mtot_cas3');
```

```python
# Generate data points
ms = np.random.normal(6.6, 0.3, 100000)

plt.figure()
plt.hist(np.log10(mtot_cas3), bins=50, alpha=0.5, label='$M$');
plt.hist(ms, bins=50, alpha=0.5, label='$M_{\star}$');


plt.legend(loc='upper left')

plt.xlabel('$\log_{10}(M/M_{\odot}$)')
```
